```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Stock Market Dataset.csv")
```

```python
df.head()
```

```python
# Rename 'Unnamed: 0' to 'ID'
df.rename(columns={'Unnamed: 0': 'ID'}, inplace=True)

# Print the first few rows to confirm the change
print("Step 1: Renamed 'Unnamed: 0' to 'ID'")
print(df.head())
```

```python

# Convert the 'Date' column to datetime format for better plotting
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Cleaning up the 'Bitcoin_Price' and 'Gold_Price' columns to convert them from string to float
df['Bitcoin_Price'] = df['Bitcoin_Price'].str.replace(',', '').astype(float)
df['Gold_Price'] = df['Gold_Price'].str.replace(',', '').astype(float)

# Selecting a subset for trend analysis
assets_to_plot = ['Natural_Gas_Price', 'Crude_oil_Price', 'Gold_Price', 'Bitcoin_Price']

# Plotting
fig, axes = plt.subplots(len(assets_to_plot), 1, figsize=(12, 15))
fig.tight_layout(pad=5.0)

for i, asset in enumerate(assets_to_plot):
    axes[i].plot(df['Date'], df[asset])
    axes[i].set_title(f'Trend of {asset.replace("_", " ")} Over Time')
    axes[i].set_xlabel('Date')
    axes[i].set_ylabel('Price')

plt.show()
```

```python
# Trend analysis with moving averages
moving_average_window = 30  # 30-day moving average

# Calculate moving averages
for asset in assets_to_plot:
    df[f'{asset}_MA'] = df[asset].rolling(window=moving_average_window).mean()

# Plotting with Moving Averages
fig, axes = plt.subplots(len(assets_to_plot), 1, figsize=(12, 15))
fig.tight_layout(pad=5.0)

for i, asset in enumerate(assets_to_plot):
    axes[i].plot(df['Date'], df[asset], label='Actual Price', alpha=0.5)
    axes[i].plot(df['Date'], df[f'{asset}_MA'], label=f'{moving_average_window}-Day Moving Average', color='blue')  # Change color here
    axes[i].set_title(f'{asset.replace("_", " ")} Trend with Moving Average')
    axes[i].set_xlabel('Date')
    axes[i].set_ylabel('Price')
    axes[i].legend()

plt.show()
```

```python
# Redefining price_columns with the correct asset price column names
price_columns = ['Natural_Gas_Price', 'Crude_oil_Price', 'Gold_Price', 'Bitcoin_Price']

# Normalization function remains the same
def normalize_prices(prices):
    return (prices / prices.iloc[0]) * 100

# Reapply normalization with the corrected variable definition
normalized_data = pd.DataFrame()
for asset in price_columns:
    normalized_data[asset] = normalize_prices(df[asset])

# Plotting the normalized prices with different colors for each asset
plt.figure(figsize=(14, 7))
colors = ['blue', 'green', 'orange', 'purple']  # Define colors for each asset

for i, asset in enumerate(normalized_data.columns):
    plt.plot(df['Date'], normalized_data[asset], label=asset.replace("_", " "), color=colors[i])

plt.title('Comparative Analysis of Asset Prices (Normalized to 100)')
plt.xlabel('Date')
plt.ylabel('Normalized Price')
plt.legend()
plt.show()
```

```python
# Calculate daily percentage changes for each asset
for asset in price_columns:
    df[f'{asset}_Change'] = df[asset].pct_change() * 100

# Selecting the daily price change columns for correlation analysis
price_change_columns = [f'{asset}_Change' for asset in price_columns]

# Compute the correlation matrix
correlation_matrix = df[price_change_columns].corr()

# Display the correlation matrix
print("Correlation Matrix of Daily Price Changes:")
print(correlation_matrix)
```
